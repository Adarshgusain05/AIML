```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sales.csv")
```

```python
data=pd.read_csv("supermarket_sales.csv")
data.info()
```

```python
plt.figure()
plt.plot(data['Quantity'],data['Total'],marker='*')
plt.title("Line Plot")
plt.xlabel("Quantity")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure()
plt.bar(data['Unit price'],data['cogs'])
plt.title("Bar Chart")
plt.xlabel("Unit price")
plt.ylabel("cogs")
plt.show()
```

```python
plt.figure()
plt.scatter(data['Quantity'],data['Unit price'])
plt.title('Scatter Plot')
plt.xlabel("Quantity")
plt.ylabel('Unit price')
plt.show()
```

```python
plt.figure()
sns.histplot(data['Quantity'],bins=6)
plt.title("Histogram")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Quantity', y='Tax 5%',data=data)
plt.title("Box Plot")
plt.xlabel("Quantity")
plt.ylabel("Tax 5%")
plt.show()
```
